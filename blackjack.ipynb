Yo creo que para hacerlo primero tiene que definir los valore de cada carta, luego ponerlos en una lista con todas las cartas.
Despues hago un random para sacar dos cartas, y luego que analice las cartas. Si tiene menos de 21 puede hacer borrow con un if statement
sino que imprima que perdio.
